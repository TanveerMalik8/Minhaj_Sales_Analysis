### State

```python

# Group and sort data
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# Define colors using Seaborn
colors = sns.color_palette('coolwarm', len(sales_state))

# Create Pie Chart
plt.figure(figsize=(7, 7))  # Adjust figure size if needed
plt.pie(sales_state['Orders'], labels=sales_state['State'], autopct='%1.1f%%', 
        colors=colors, startangle=140, wedgeprops={'edgecolor': 'black'})

# Add Title
plt.title("Top 10 States by Orders", fontsize=16, fontweight='bold')

# Show Pie Chart
plt.show()
```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

# Set figure size
sns.set(rc={'figure.figsize': (15, 5)})

# Create Line Chart
sns.lineplot(data=sales_state, x='State', y='Amount', marker='o', linewidth=2, markersize=8, color='b')

# Add Title and Labels
plt.title("Top 10 States by Sales Amount", fontsize=14, fontweight='bold')
plt.xlabel("State")
plt.ylabel("Total Sales Amount")

# Show Grid
plt.grid(True, linestyle='--', alpha=0.6)

# Show Plot
plt.show()
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*


### Marital Status

```python

marital_counts = df['Marital_Status'].value_counts()

# Define colors using Seaborn palette
colors = sns.color_palette("pastel", len(marital_counts))

# Create a Pie Chart (Donut Chart)
plt.figure(figsize=(4, 4))
wedges, texts, autotexts = plt.pie(marital_counts.values, labels=marital_counts.index, autopct='%1.1f%%',
                                   colors=colors, startangle=140, wedgeprops={'edgecolor': 'black'})

# Add center circle to make it a Donut Chart
plt.gca().add_artist(plt.Circle((0, 0), 0.5, fc='white'))

# Title
plt.title("Marital Status Distribution", fontsize=14, fontweight='bold')

# Show Plot
plt.show()
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(3,6)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount',  palette='cool',hue='Marital_Status')
```

### Occupation

```python
# Set figure size
sns.set(rc={'figure.figsize': (10, 5)})

# Create countplot and apply palette correctly
ax = sns.countplot(data=df, x='Occupation', hue='Occupation', palette='cool', legend=False)

# Rotate x-axis labels if necessary
plt.xticks(rotation=45)

# Show plot
plt.show()
```

```python
# Group and sort data
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Set figure size
sns.set(rc={'figure.figsize': (10, 6)})

# Create a horizontal bar chart
sns.barplot(data=sales_state, y='Occupation', x='Amount', hue= 'Amount' , palette='viridis')

# Add title and labels
plt.title("Total Sales Amount by Occupation", fontsize=14, fontweight='bold')
plt.xlabel("Total Sales Amount")
plt.ylabel("Occupation")

# Show the plot
plt.show()
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

### Product Category

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category' ,hue='Product_Category' ,palette='viridis')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Group and sort data
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

# Set figure size
sns.set(rc={'figure.figsize': (12, 6)})

# Create a histogram using seaborn histplot
sns.histplot(data=sales_state, x='Amount', bins=10, kde=True, color='blue')

# Add title and labels
plt.title("Distribution of Sales Amount by Product Category (Top 10)", fontsize=14, fontweight='bold')
plt.xlabel("Sales Amount")
plt.ylabel("Frequency")

# Show the plot
plt.show()
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*

```python
# Group and sort data
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# Create a categorical plot (bar type)
sns.catplot(data=sales_state, x='Product_ID', y='Orders', kind='bar',hue='Product_ID',palette='coolwarm', height=5, aspect=2)

# Add title and labels
plt.title("Total Orders for Top 10 Products", fontsize=14, fontweight='bold')
plt.xlabel("Product ID")
plt.ylabel("Number of Orders")

# Rotate x-axis labels for readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
# Group and sort data
top_products = df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).reset_index()

# Create the figure
plt.figure(figsize=(12, 7))

# Create a line plot
sns.lineplot(data=top_products, x='Product_ID', y='Orders', marker='o', linewidth=2, color='b')

# Add title and labels
plt.title("Top 10 Most Sold Products", fontsize=14, fontweight='bold')
plt.xlabel("Product ID")
plt.ylabel("Number of Orders")

# Rotate x-axis labels for readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
